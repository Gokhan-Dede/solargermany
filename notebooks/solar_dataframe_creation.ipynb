```python
import pandas as pd
```

```python
# Load the Parquet file
data = pd.read_parquet('./solar_panels/EinheitenSolar.parquet')
```

```python
data.rename(columns={
    'EinheitMastrNummer': 'UnitMasterNumber',
    'DatumLetzteAktualisierung': 'LastUpdateDate',
    'LokationMaStRNummer': 'LocationMaStRNumber',
    'NetzbetreiberpruefungStatus': 'GridOperatorCheckStatus',
    'AnlagenbetreiberMastrNummer': 'FacilityOperatorMasterNumber',
    'Land': 'Country',
    'Bundesland': 'State',
    'Landkreis': 'Administrative Region',
    'Gemeinde': 'City',
    'Gemeindeschluessel': 'MunicipalityCode',
    'Postleitzahl': 'PostalCode',
    'Ort': 'Site',
    'Registrierungsdatum': 'RegistrationDate',
    'Inbetriebnahmedatum': 'CommissioningDate',
    'EinheitSystemstatus': 'UnitSystemStatus',
    'EinheitBetriebsstatus': 'UnitOperationalStatus',
    'NichtVorhandenInMigriertenEinheiten': 'NotPresentInMigratedUnits',
    'NameStromerzeugungseinheit': 'PowerGenerationUnitName',
    'Weic_nv': 'Weic_Nv',
    'Kraftwerksnummer_nv': 'PowerPlantNumber_Nv',
    'Energietraeger': 'EnergyCarrier',
    'Bruttoleistung': 'GrossPower',
    'Nettonennleistung': 'NetRatedPower',
    'FernsteuerbarkeitNb': 'RemoteControllabilityNb',
    'Einspeisungsart': 'FeedInType',
    'ZugeordneteWirkleistungWechselrichter': 'AssignedActivePowerInverter',
    'AnzahlModule': 'NumberOfModules',
    'Lage': 'Location',
    'Leistungsbegrenzung': 'PowerLimitation',
    'EinheitlicheAusrichtungUndNeigungswinkel': 'UniformOrientationAndTiltAngle',
    'Hauptausrichtung': 'MainOrientation',
    'HauptausrichtungNeigungswinkel': 'MainOrientationTiltAngle',
    'Nutzungsbereich': 'UsageArea',
    'EegMaStRNummer': 'EegMaStRNumber',
    'Nebenausrichtung': 'SecondaryOrientation',
    'NebenausrichtungNeigungswinkel': 'SecondaryOrientationTiltAngle',
    'Strasse': 'Street',
    'StrasseNichtGefunden': 'StreetNotFound',
    'Hausnummer_nv': 'HouseNumber_Nv',
    'HausnummerNichtGefunden': 'HouseNumberNotFound',
    'Laengengrad': 'Longitude',
    'Breitengrad': 'Latitude',
    'DatumBeginnVoruebergehendeStilllegung': 'TemporaryShutdownStartDate',
    'Hausnummer': 'HouseNumber',
    'Adresszusatz': 'AddressSupplement',
    'FernsteuerbarkeitDv': 'RemoteControllabilityDv',
    'NetzbetreiberpruefungDatum': 'GridOperatorCheckDate',
    'DatumDesBetreiberwechsels': 'OperatorChangeDate',
    'DatumRegistrierungDesBetreiberwechsels': 'OperatorChangeRegistrationDate',
    'Gemarkung': 'Parish',
    'FlurFlurstuecknummern': 'ParcelPlotNumber',
    'AnschlussAnHoechstOderHochSpannung': 'ConnectionToHighVoltage',
    'InAnspruchGenommeneFlaeche': 'ClaimedArea',
    'ArtDerFlaecheIds': 'AreaTypeIds',
    'DatumEndgueltigeStilllegung': 'FinalDecommissioningDate',
    'InAnspruchGenommeneAckerflaeche': 'ClaimedAgriculturalArea',
    'GenMastrNummer': 'GenMasterNumber',
    'Einsatzverantwortlicher': 'DeploymentResponsible',
    'WeicDisplayName': 'WeicDisplayName',
    'DatumWiederaufnahmeBetrieb': 'OperationResumptionDate',
    'GeplantesInbetriebnahmedatum': 'PlannedCommissioningDate',
    'Buergerenergie': 'CitizenEnergy',
    'Kraftwerksnummer': 'PowerPlantNumber',
    'Weic': 'Weic'
}, inplace=True)

# Print the updated column names
print(data.columns)
```

```python
geographical_based_features=["Country", "State", "Administrative Region","City"]


energy_related_features=["GrossPower",
                         "MainOrientation",
                         "MainOrientationTiltAngle",
                         "NetRatedPower",
                         "FeedInType",
                         "AssignedActivePowerInverter",
                         "NumberOfModules",
                         "Location"]

time_based_features=["RegistrationDate",
                     "CommissioningDate"]

operational_details=["UnitOperationalStatus"]

categorical_features=["UniformOrientationAndTiltAngle"]
```

```python
df = pd.concat([data[geographical_based_features],
                data[energy_related_features],
                data[time_based_features],
                data[operational_details],
                data[categorical_features]], axis=1)


# Display the structure of the combined DataFrame
df.head()
```

```python
df["Country"].replace("84", "Germany", inplace=True)
```

```python
state_mapping = {
    "1400":"Brandenburg",
    "1401":"Berlin",
    "1402":"Baden-Württemberg",
    "1403":"Bayern",
    "1405":"Hessen",
    "1404":"Bremen",
    "1406":"Hamburg",
    "1407":"Mecklenburg-Vorpommern",
    "1408":"Niedersachsen",
    "1409":"Nordrhein-Westfalen",
    "1410":"Rheinland-Pfalz",
    "1411":"Schleswig-Holstein",
    "1412":"Saarland",
    "1413":"Sachsen",
    "1414":"Sachsen-Anhalt",
    "1415":"Thüringen",
    "1416":"Ausschließliche Wirtschaftszone"
}
```

```python
df["State"] = df["State"].replace(state_mapping)
```

```python
orientation_mapping={"695": "North",
                     "696": "North-East",
                     "697": "East",
                     "698": "South-East",
                     "699": "South",
                     "700": "South-West",
                     "701": "West",
                     "702": "North-West",
                     "703": "Adjusted",
                     "704": "East-West"
                    }
```

```python
df["MainOrientation"] = df["MainOrientation"].replace(orientation_mapping)
```

```python
orientation_angle_mapping={"806": "Facade-integrated",
                           "807": "> 60 Grad",
                           "808": "40 - 60 Grad",
                           "809": "20 - 40 Grad",
                           "810": "< 20 Grad",
                           "811": "Adjusted"}
```

```python
df["MainOrientationTiltAngle"]=df["MainOrientationTiltAngle"].replace(orientation_angle_mapping)
```

```python
feedintype_mapping={"688":"Full Feed-in",
                    "689":"Partial Feed-in"}
```

```python
df["FeedInType"]=df["FeedInType"].replace(feedintype_mapping)
```

```python
location_mapping={"852":"Open land/field",
                  "853":"Structural installations (house roof, buildings, and facades)",
                  "2484":" Other structural installations",
                  "2961":"Plug-and-play solar system (balcony power plant)",
                  "3002":"Water bodies",
                  "3058":"Large parking lot"}
```

```python
df["Location"]=df["Location"].replace(location_mapping)
```

```python
operational_mapping={"35":"In operation",
                     "37":"Temporarily shut down",
                     "38":"Permanently shut down"}
```

```python
df["UnitOperationalStatus"]=df["UnitOperationalStatus"].replace(operational_mapping)
```

```python
df.to_csv('solar_dataframe.csv', index=False)
```
